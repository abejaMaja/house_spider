```python
import pandas as pd
import numpy as np
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
pd.options.display.max_rows
pd.set_option('display.max_colwidth', -1)
```

```python
PATH = '~/Projekty/Dane/output.csv'
data = pd.read_csv(PATH)
data.head()
```

```python
def make_clickable(val):
    return '<a href="{}">{}</a>'.format(val,val)

data['link'] = data['link'].apply(make_clickable)
```

```python
data['price'] = data['price'].apply(lambda x: x.replace('zł',''))
data['price'] = data['price'].apply(lambda x: x.replace(' ',''))
data['price'] = data['price'].apply(lambda x: x.replace('~',''))
data['price'] = data['price'].astype(str).astype(int)
```

## **Strzyża**

```python
s = data.loc[data['subtitle'].str.contains('strzy', case=False)] 
HTML(s.loc[s['price'] <= 1500000].sort_values(by='price',ascending=1).to_html(escape=False)) 
#HTML(s.to_html(escape=False))
```
